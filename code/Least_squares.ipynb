```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.1)
y = 2*np.cos(x) + (1/2)*x
```

```python
plt.plot(x, y)
```

```python
# add random number
y = 2*np.cos(x) + (1/2)*x + np.random.normal(0.0, 1.0, len(x))
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# generate dataset
n = 2
X = np.zeros((len(x), n), float)

X[:, 0] = np.cos(x)
X[:, 1] = x + np.random.normal(0.0, 0.5, len(x))
```

```python
# estimate Î¸
theta, residuals, rank, s = np.linalg.lstsq(X, y, rcond=None)
```
